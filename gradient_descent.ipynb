### Gradient Descent

```python
import random
import sys

learning_rate = 0.1  # default learning rate
num_iteration = 100  # default number of iteration
count = 0  # counter
x = random.uniform(-sys.maxsize, sys.maxsize)  # Set random value for x
def df(i): return 6 * i + 2  # Gradient descent


print("Initial value of x: ", x)

while count < num_iteration:
    x = x - learning_rate * df(x)
    count += 1
    print("#Iteration: ", count)
    print("value of x: ", x)
print("Global minimum of the function: ", x)
```
